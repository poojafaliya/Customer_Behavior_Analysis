```python
import pandas as pd
```

```python
df= pd.read_csv("/content/customer_shopping_behavior.csv")
```

```python
df
```

```python
df.info()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python

df.info()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a mew column age_group
labels = ['young adult','adult','middile-adge','senior']
df['age_group'] = pd.qcut(df['age'],q = 4,labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'fortnightly': 14,
    'weekly': 7,
    'monthly': 30,
    'quarterly': 90,
    'bi-weekly': 14,
    'annually': 365,
    'every 3 months': 90
}

df['purchase_frequancy_days'] = df['frequency_of_purchases'].str.lower().map(frequency_mapping)
```

```python
df[['purchase_frequancy_days','frequency_of_purchases']].head(10)
```
